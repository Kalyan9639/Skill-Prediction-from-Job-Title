```python
import pandas as pd

df = pd.read_csv('extracted_job_data.csv')
df.head()
```

```python
df[df['Source'] == 'linkedin_jobs.json'].head()
```

```python
df1 = df.dropna()
df1.isna().sum()
```

```python
df1.shape
```

```python
df1 = df1.drop("Source",axis=1)
df1.head()
```

```python
from langchain_ollama import OllamaEmbeddings

embeddings = OllamaEmbeddings(model="nomic-embed-text")
```

```python
embeddings.embed_query("eBay, Amazon")
```

```python
df1.shape
```

```python
df2 = df1.copy()

df2['job_title_vector'] = df1['Job Title'].apply(lambda x: embeddings.embed_documents([x])[0])
df2['job_description_vector'] = df1['Job Description'].apply(lambda x: embeddings.embed_documents([x])[0])
df2['skills_vector'] = df1['Skills'].apply(lambda x: embeddings.embed_documents([x])[0])
```

```python
df2.head()
```

```python
df2.to_csv('job_embeddings.csv', index=False)
```
