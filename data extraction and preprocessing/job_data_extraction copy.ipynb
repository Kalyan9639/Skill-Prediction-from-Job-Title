```python
import requests

url = "https://upwork-jobs-api2.p.rapidapi.com/active-freelance-1h"

querystring = {"limit":"100","category":"Ai Agents"}

headers = {
	"x-rapidapi-key": "c8487cf183msh57bb29c50fe7914p1360d9jsn2ba50f0ec64b",
	"x-rapidapi-host": "upwork-jobs-api2.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```

```python
import json
from IPython.display import JSON

# json_obj = JSON(response.json())
json_str = json.dumps(response.json(), indent=4)
```

```python
from datetime import datetime

date_time = datetime.now()
if date_time.hour > 12:
    hour = date_time.hour - 12
    measurement = "P.M."
else:
    hour = date_time.hour
    measurement = "A.M."

with open(f"{date_time.date()} {hour}{measurement}.json", "w") as file:
    file.write(json_str)
```

### Using Linkedin API

```python
import requests

url = "https://linkedin-job-search-api.p.rapidapi.com/active-jb-1h"

querystring = {"offset":"0","title_filter":"Data Engineer","location_filter":"India"}

headers = {
	"x-rapidapi-key": "c8487cf183msh57bb29c50fe7914p1360d9jsn2ba50f0ec64b",
	"x-rapidapi-host": "linkedin-job-search-api.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```

```python
linkedin_json_str = json.dumps(response.json(), indent=4)
```

```python
from datetime import datetime

date_time = datetime.now()
if date_time.hour > 12:
    hour = date_time.hour - 12
    measurement = "P.M"
else:
    hour = date_time.hour
    measurement = "A.M"

with open(f"{date_time.date()} {hour}{measurement}_linkedin.txt", "w") as file:
    file.write(linkedin_json_str)
```
